**4.8 A derivada de $f(x) = 1/(1 − 3x²)$ é dada por:**

$f'(x) = \frac{6x}{(1-3x²)²}$

Avalie $f'(0.577)$, com três e quatro algarismos de truncamento

```python
D = []
I = []

for i in range(1, 1000):
    x = i/1000 
    n = (6*x)
    d = (1 - 3 * np.power(x, 2))
    d = np.power(d, 2)
    y = n/d
    D.append(x)
    t = np.trunc( y*10**4/10**4 )
    I.append(t)
    

D = np.array(D)
I = np.array(I)

print(np.min(I))

plt.plot(D, I)
plt.show()

```

**4.9 Calcule o polinômio**

```python
### Questão 4.9 a) ###

def erroRelativo(va, vt): # va = vavlor absoluto, vt = valor truncado
    return (va - vt)/vt
    

X = []
Y = []
Ytrunc = []
E = []

for i in range(1, 101):
    
    x = i/100 + 1
    X.append(x)
    y = x**3 - 7*x**2 + 8*x - 0.35
    Y.append(y)
    t = np.trunc( y*10**3/10**3 )
    Ytrunc.append(t)
    E.append(erroRelativo(y, t))
    

X = np.array(X)
Y = np.array(Y)
erro = np.array(E)


Ytrunc = np.array(Ytrunc)
plt.title('Erro Relativo')
plt.plot(X, Y, color='#17a589', label='Absoluto')
plt.plot(X, Ytrunc, color='red', label='Truncado')
plt.legend()
```

**4.11 A expansão em séria de MacLaurin para $\cos{x}$ é:** <br /><br />

$ \cos{x} = 1 - \frac{x²}{2} + \frac{x⁴}{4!} - \frac{x⁶}{6!} + \frac{x⁸}{8!} - \ldots$ <br />



```python
import math
import numpy as np
import matplotlib.pyplot as plt

def fatorial(n):
    fat = 1
    i = 2
    while i <= n:
        fat = fat*i
        i = i + 1

    return fat

pi = math.pi
sequencia = 1
positivo = False
i = 2

Absoluto = []
Estimado = []
X = [1]

sizeSequence = 20

while ( i < sizeSequence ):
    
    f = fatorial(i)
    
    if (positivo):
        sequencia += ((pi/4)**i)/f
        positivo = False
        i+=2
    else:
        sequencia -= ((pi/4)**i)/f
        positivo = True
        i+=2
    
    Absoluto.append(math.cos(pi/4))
    Estimado.append(sequencia)
    X.append(i)
    

print("Absoluto: " + str(math.cos(pi/4)) + "\n" + "Estimado: " + str(sequencia) )

Absoluto = np.array(Absoluto)
Estimado = np.array(Estimado)
X = np.array(X)

plt.title('MacLaurin')
plt.plot(Absoluto, color='#17a589', label='Absoluto')
plt.plot(Estimado, color='red', label='MacLaurin')
plt.legend()
```

Quanto maior o tamanho da sequencia definido em **sizeSequence** mais aproximado será o valor estimado de $cos(\frac{\pi}{4})$
